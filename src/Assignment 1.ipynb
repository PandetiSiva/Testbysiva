Random team Generation assignment based on the number of teams provided in a variable:
Read a file which has student names in a class, copy this file data in a list with list comprehension concept. Number of teams is provided in a variable. This variable can be modified as per the need which plays crucial role in deciding how many people to be added in a team.
With the help of Random Sample (without replacement) and loop, random teams are created each time when you run the sam code again and again based on number of teams provided in a variable.

```python
import csv
import random

with open('C://Users/sgsup/Desktop/CSC605 file1.tsv', newline='') as f:
    reader = csv.reader(f,delimiter='\t')
    data = list(reader)
students=[]
[students.append(data[a][0]) for a in range(1,len(data))]

number_of_students = len(students)
number_of_teams = 6
while number_of_students > 0 and number_of_teams > 0:
    team = random.sample(students, int(number_of_students/number_of_teams))
    for x in team:
        students.remove(x)
    number_of_students -= int(number_of_students/number_of_teams)
    number_of_teams -= 1
    print(team)
```

```python
import csv
import random

with open('C://Users/sgsup/Desktop/CSC605 file1.tsv', newline='') as f:
    reader = csv.reader(f,delimiter='\t')
    data = list(reader)
students=[]
[students.append(data[a][0]) for a in range(1,len(data))]

number_of_students = len(students)
number_of_teams = 10; team_number = 0
if number_of_teams > number_of_students:
    raise Exception("number of teams are greater than number of students in a class")
while number_of_students > 0:
    sample_size = int(number_of_students/number_of_teams)
    team = random.sample(students, sample_size)
    team_number += 1
    for std in team:
        students.remove(std)
    number_of_students = number_of_students - sample_size; number_of_teams = number_of_teams - 1
    print(('Team Number: {}\n' + 
            '{}').format(team_number,team))
```
